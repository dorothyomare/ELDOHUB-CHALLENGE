```python
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize
import requests
pd.set_option('max_rows', 99999)
pd.set_option('max_colwidth', 400)
pd.describe_option('max_colwidth')
```

```python
URL = 'https://raw.githubusercontent.com/onaio/ona-tech/master/data/water_points.json'
#reading the url to data frame
df = pd.read_json(URL)
print(df)
```

```python
data = pd.DataFrame(df)
data.head()
```

```python
essential_attr = data[['_id','communities_villages', 'water_point_condition']]
essential_attr.head(10)
```

# Number of functioning waterpoints

```python
essential_attr.isnull().sum()
```

```python
def WaterFunctional(essential_attr):
    okay= essential_attr.water_point_condition.value_counts()
    return okay
WaterFunctional(essential_attr)
```

The number of functioning water points is 590

**The number of water points per community**,

# The number of water points per community,

```python
def WaterPerCommunity(essential_attr):
    summary =essential_attr.groupby(essential_attr["communities_villages"])['water_point_condition'].value_counts()
    return summary
    
WaterPerCommunity(essential_attr)
```

#The rank for each community by the percentage of broken water points.

```python
def PercentageBrocken(essential_attr):
    counts =essential_attr.groupby(essential_attr["communities_villages"])['water_point_condition'].value_counts('broken')
    return counts
PercentageBrocken(essential_attr) 
```
